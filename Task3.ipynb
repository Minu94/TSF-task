## Prediction Using Decision Tree Algorithm

### Task: For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Overview

```python
iris_df=pd.read_csv("Iris.csv")
iris_df.head()
```

```python
iris_df=iris_df.drop("Id",axis=1)
```

```python
iris_df.isna().sum()
```

## Data Visualization

```python
sns.set_style("darkgrid")
sns.pairplot(iris_df,hue='Species',palette='hls');
```

## Preparing dataset for trainig

```python
x=iris_df.iloc[:,:-1]
y=iris_df.iloc[:,-1]
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Training Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print('accuracy: ',accuracy_score(y_pred,y_test))
```

## Decision Tree Visualization

```python
plt.figure(figsize =( 50,30))
plt.style.use('grayscale') 
plot_tree(classifier,rounded=True,filled=True,feature_names=x.columns);
```
